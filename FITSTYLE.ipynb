# ***USING OPENPOSE TO EXTRACT THE KEYPOINTS FROM IMAGES***

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import cv2
import numpy as np
import os
from google.colab import drive

# Mount Google Drive (if your images are there)
drive.mount('/content/drive')

# Load model files
protoFile = "models/pose/body_25/pose_deploy.prototxt"
weightsFile = "models/pose/body_25/pose_iter_584000.caffemodel"

net = cv2.dnn.readNetFromCaffe(protoFile, weightsFile)

# Folder with your images
image_folder = "/content/drive/MyDrive/fashion_images"  # <-- update if different
output = {}

def extract_keypoints(image_path):
    frame = cv2.imread(image_path)
    if frame is None:
        return None
    frameWidth, frameHeight = frame.shape[1], frame.shape[0]
    inpBlob = cv2.dnn.blobFromImage(frame, 1.0 / 255, (368, 368),
                                    (0, 0, 0), swapRB=False, crop=False)

    net.setInput(inpBlob)
    output_data = net.forward()

    H, W = output_data.shape[2], output_data.shape[3]

    points = []
    threshold = 0.1
    for i in range(25):  # BODY_25 model has 25 keypoints
        probMap = output_data[0, i, :, :]
        minVal, prob, minLoc, point = cv2.minMaxLoc(probMap)

        x = (frameWidth * point[0]) / W
        y = (frameHeight * point[1]) / H

        if prob > threshold:
            points.append((int(x), int(y)))
        else:
            points.append(None)
    return points

# Process all images
results = {}
for filename in os.listdir(image_folder):
    if filename.endswith((".jpg", ".png")):
        img_path = os.path.join(image_folder, filename)
        keypoints = extract_keypoints(img_path)
        results[filename] = keypoints

print(f"Processed {len(results)} images.")
```

```python
import pandas as pd

def euclidean(p1, p2):
    return ((p1[0]-p2[0])**2 + (p1[1]-p2[1])**2) ** 0.5

measurement_data = []

for filename, keypoints in results.items():
    if not keypoints:
        continue

    try:
        bust = euclidean(keypoints[2], keypoints[5]) if keypoints[2] and keypoints[5] else None
        waist = euclidean(keypoints[8], keypoints[11]) if keypoints[8] and keypoints[11] else None
        hip = euclidean(keypoints[9], keypoints[12]) if keypoints[9] and keypoints[12] else None

        measurement_data.append({
            "filename": filename,
            "bust": bust,
            "waist": waist,
            "hip": hip
        })
    except:
        continue

df_measurements = pd.DataFrame(measurement_data)
df_measurements.to_csv("/content/drive/MyDrive/body_measurements.csv", index=False)
```

# ***KNN IMPLEMENTATION FOR RECOMMENDATION SYSTEM***

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import NearestNeighbors
import cv2
import matplotlib.pyplot as plt
import os

# Load dataset
df = pd.read_csv("/content/drive/MyDrive/body_measurements_clean.csv")
features = df[["bust", "waist", "hip"]].values
filenames = df["filename"].values

# Fit KNN
knn = NearestNeighbors(n_neighbors=5)
knn.fit(features)

# Function to recommend outfits and show images
def recommend_and_display(user_bust, user_waist, user_hip, image_folder, k=5):
    input_vector = np.array([[user_bust, user_waist, user_hip]])
    distances, indices = knn.kneighbors(input_vector, n_neighbors=k)

    plt.figure(figsize=(15, 5))
    for i, idx in enumerate(indices[0]):
        img_path = os.path.join(image_folder, filenames[idx])
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plt.subplot(1, k, i+1)
        plt.imshow(img)
        plt.axis('off')
        plt.title(f"Rank {i+1}")
    plt.show()

# Example usage
recommend_and_display(user_bust=90, user_waist=70, user_hip=95,
                      image_folder="/content/drive/MyDrive/fashion_images")
```

```python
import numpy as np

def get_measurements_from_image(img_path):
    frame = cv2.imread(img_path)
    frameWidth = frame.shape[1]
    frameHeight = frame.shape[0]

    inpBlob = cv2.dnn.blobFromImage(frame, 1.0 / 255, (368, 368),
                                    (0, 0, 0), swapRB=False, crop=False)

    net.setInput(inpBlob)
    output = net.forward()

    H, W = output.shape[2], output.shape[3]
    points = []

    for i in range(25):
        probMap = output[0, i, :, :]
        minVal, prob, minLoc, point = cv2.minMaxLoc(probMap)

        x = (frameWidth * point[0]) / W
        y = (frameHeight * point[1]) / H

        if prob > 0.1:
            points.append((int(x), int(y)))
        else:
            points.append(None)

    # Measure distances if keypoints exist
    bust = np.linalg.norm(np.array(points[2]) - np.array(points[5])) if points[2] and points[5] else None
    waist = np.linalg.norm(np.array(points[8]) - np.array(points[11])) if points[8] and points[11] else None
    hip = np.linalg.norm(np.array(points[9]) - np.array(points[12])) if points[9] and points[12] else None

    return bust, waist, hip

# Example
user_img = list(uploaded.keys())[0]
bust, waist, hip = get_measurements_from_image(user_img)
print(f"User measurements → Bust: {bust}, Waist: {waist}, Hip: {hip}")
```
